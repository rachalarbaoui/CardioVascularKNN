```python
df = pd.read_csv("/home/cardiovascular_diseases_dv3.csv")
df.head(10)
```

```python
df.info()
```

```python
X = df.drop("CARDIO_DISEASE", axis = 1)
Y = df['CARDIO_DISEASE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42,shuffle = True)
```

```python
sc_X=StandardScaler()
X_train[['AGE','HEIGHT','WEIGHT','AP_HIGH','AP_LOW']]=sc_X.fit_transform(X_train[['AGE','HEIGHT','WEIGHT','AP_HIGH','AP_LOW']])
X_test[['AGE','HEIGHT','WEIGHT','AP_HIGH','AP_LOW']] = sc_X.transform(X_test[['AGE','HEIGHT','WEIGHT','AP_HIGH','AP_LOW']])
```

```python
print(len(y_test))
```

```python
X_train.head(10)
```

```python
model=KNeighborsClassifier(n_neighbors=117,p=2,metric='euclidean')
```

```python
model.fit(X_train,y_train)
```

```python
prediction_test = model.predict(X_test)
```

```python
print ("Accuracy = ", metrics.accuracy_score(y_test, prediction_test))
```

```python
# définir les valeurs de paramètres à rechercher

k_range = list(range(3, 200))
print(k_range)
```

```python
model2=KNeighborsClassifier(p=2,metric='euclidean')
```

```python
param_grid = dict(n_neighbors=k_range)
print(param_grid)
```

```python
# instancier la grille
grid = GridSearchCV(model2, param_grid, cv=5, scoring='accuracy',verbose=2)
```

```python
# adapter la grille avec des données
grid.fit(X_train,y_train)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
#après avoir essayé ces paramètres, cela montre que ces paramètres ne sont pas les meilleurs paramètres, nous devons donc les rechercher manuellement
```

```python
grid.cv_results_
```

```python
#après la recherche, nous avons trouvé que les meilleurs paramètres sont k = 106
```

```python
model2=KNeighborsClassifier(p=2,metric='euclidean',n_neighbors=106)
```

```python
model2.fit(X_train, y_train)
```

```python
prediction_test2 = model2.predict(X_test)
```

```python
print ("Accuracy = ", metrics.accuracy_score(y_test, prediction_test2))
```
